```
def ifelse(a):
  if a == 0:
    return 1
  else:
    return a
```

```
def rms_norm(v_x):
  n_rms = np.sqrt(np.sum(v_x ** 2)/(len(v_x) - 1))
  return v_x/n_rms
```

```
data = data.applymap(ifelse)
```
